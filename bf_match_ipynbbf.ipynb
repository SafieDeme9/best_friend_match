```python
def distance_couple(row1,row2) :
  distance = 0
  for i in range(0,len(row1)):
    distance = distance + (row1[i] - row2[i])**2
  return distance

```

```python
df_list = data.values.tolist()
print(df_list)
```

```python
def distance_allcouples(listoflists) :
  dict_couple = {}
  for i in range(0, len(listoflists)) :
    for j in range(0, len(listoflists)) :
      if distance_couple(listoflists[i],listoflists[j]) != 0 :
        dict_couple[i,j] = distance_couple(listoflists[i],listoflists[j])
  return dict_couple
```

```python
def create_final_dict(sorted_dict):
  sorted_dict = dict(sorted(distance_allcouples(df_list).items(), key=lambda x: x[1]))
  final_dict = {}
  existing_index = []
  for i in sorted_dict.keys():
    index1 = i[0]
    index2 = i[1]
    if i[0] not in existing_index and i[1] not in existing_index :
      final_dict[i] = sorted_dict[i]
      existing_index.append(index1)
      existing_index.append(index2)
  return final_dict
```

```python
def listofthecouples(final_dict):
  names = df["Nom et Pr√©nom(s)"]
  couples = []
  for i in final_dict.keys():
    name1 = i[0]
    name2 = i[1]
    couples.append((names[name1], names[name2]))
  return couples
```

```python
sorted_dict = dict(sorted(distance_allcouples(df_list).items(), key=lambda x: x[1]))
couplesanddistances = create_final_dict(sorted_dict)
bestofriendos = listofthecouples(couplesanddistances)
df_bf = pd.DataFrame(bestofriendos)
df_bf.to_csv('results.csv', index= False)
```
